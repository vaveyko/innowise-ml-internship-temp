```python
from pandas import read_csv
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
# load data after ETL level
data = read_csv("../data/preprocessed/sales_train_preprocessed.csv")
# data = read_csv("../data/preprocessed/sales_train_preprocessed_long.csv")
data.head(3)
```

```python
len(data)
```

```python
model = RandomForestRegressor(
        n_estimators=150,
        max_depth=10,
        max_features="log2",
        min_samples_split=40,
        verbose=5,
        oob_score=True
    )
```

```python
X_train = data.drop(columns=['item_cnt_month', 'date_block_num'])
y_train = data[['item_cnt_month']]
```

```python
model.fit(X_train, y_train['item_cnt_month'].to_list())
```

```python
# load test data
data_test = read_csv("../data/preprocessed/test_preprocessed.csv")
data_test.head(3)
```

```python
answer = data_test[['ID']]
answer['ID'] = answer['ID'].astype(int)
answer['item_cnt_month'] = model.predict(data_test.drop(columns='ID'))
# answer['item_cnt_month'] = model.predict(data_test_for_model.drop(columns='ID'))
# answer['item_cnt_month'] = answer['item_cnt_month'].apply(lambda x: x if x != 0 else 1)
# target postprocessing
answer['item_cnt_month'] = np.floor(answer['item_cnt_month'])
answer
```

```python
answer.to_csv("../data/preprocessed/sales_test_answer.csv", index=False)
```
